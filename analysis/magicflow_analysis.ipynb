```python
import pandas as pd


# magic flow data for tuesday
df = pd.read_json('/Users/oreogundipe/lab/fusion/analysis/data.json')

# get all the magicflow contexts
magicFlow_contexts = pd.DataFrame(df["2022-12-10"]["productivityMetrics"]["contexts"])
magicFlow_contexts = magicFlow_contexts[["app", "timestamp", "endDate", "focus", "work","categories", "duration"]]

# get all the magicflow events
magicFlow_rawEvents = pd.DataFrame(df["2022-12-10"]["window"]['events'])
magicFlow_rawEvents
```

```python
# neurosity focus & calm data for tuesday
focus_file_list = [
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1669217984.json",
    # day of recording for fellowship
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1669220847.json"
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1670719990.json",
    # last night work session
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1670721205.json",
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1670721823.json",
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1670724035.json",
    "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/focus_1670725005.json"
]
focus_df_list = [] # an empty list to store the data frames

for file in focus_file_list:
    data = pd.read_json(file) # read data frame from json file
    focus_df_list.append(data) # append the data frame to the list

df_neurosity_focus = pd.concat(focus_df_list) # concatenate all the data frames in the list.

df_neurosity_focus
```

```python
# read power spectrum data
powerSpectrum_file_list = [
    # last night work session
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/powerByBand_1670721205.json",
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/powerByBand_1670721823.json",
    # "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/powerByBand_1670724035.json",
    "/Users/oreogundipe/lab/fusion/neuroscripts/logger/archive/data/powerByBand_1670725005.json"
]

powerSpectrum_df_list = [] # an empty list to store the data frames

for file in powerSpectrum_file_list:
    data = pd.read_json(file) # read data frame from json file
    powerSpectrum_df_list.append(data) # append the data frame to the list

df_neurosity_powerSpectrum = pd.concat(powerSpectrum_df_list) # concatenate all the data frames in the list.

df_neurosity_powerSpectrum
```

```python
# ensure datetime columns are in utc datetime format
df_neurosity_focus['timestamp']= pd.to_datetime(df_neurosity_focus['timestamp'], utc=True)

df_neurosity_powerSpectrum['unixTimestamp']= pd.to_datetime(df_neurosity_powerSpectrum['unixTimestamp'], unit='s', utc=True)

magicFlow_contexts['timestamp']= pd.to_datetime(magicFlow_contexts['timestamp'], utc=True)
magicFlow_contexts['endDate']= pd.to_datetime(magicFlow_contexts['endDate'], utc=True)

magicFlow_rawEvents['timestamp']= pd.to_datetime(magicFlow_rawEvents['timestamp'], utc=True)
magicFlow_rawEvents['endDate']= pd.to_datetime(magicFlow_rawEvents['endDate'], utc=True)

# df_neurosity_powerSpectrum.dtypes
```

```python
import pandas as pd
import numpy as np
import requests
from datetime import datetime, timedelta
import dateutil.parser as dp
```

```python
class Oura:
    def __init__(self):
        self.url = 'https://api.ouraring.com/v2/usercollection/heartrate'
        token = 'NZDGAC4APTALZMRXO7FMBPPWNRIQDRE3'
        self.headers = { 
            'Authorization': f'Bearer {token}' 
        }

    def getHeartRate(self, end_time= datetime.utcnow(), duration=24*7):
        start_time = end_time - timedelta(hours=duration)
        params = {
            'start_datetime': start_time, 
            'end_datetime': end_time
        }
        response = requests.request('GET', self.url, headers=self.headers, params=params) 

        return response.json()

# end_time = datetime(2022,3,1)

hrpd = pd.DataFrame(Oura().getHeartRate()['data'])
```

```python
pd.set_option('max_rows', None)

hrpd['local_timestamp'] = hrpd['timestamp'].apply(lambda x: dp.parse(x) - timedelta(hours=7))
hrpd[['local_timestamp','bpm']].sort_values(by='local_timestamp', ascending=False)
```
